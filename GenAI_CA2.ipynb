```python
#Q5
import pandas as pd
import numpy as np

def generate_model_data(num_records=100):
    np.random.seed(0)  # For reproducibility
    data = {
        'Fever': np.random.randint(0, 11, size=num_records),
        'Cold': np.random.randint(0, 11, size=num_records),
        'Shivering': np.random.randint(0, 11, size=num_records),
        'Weight Loss': np.random.randint(0, 11, size=num_records),
    }
    df = pd.DataFrame(data)
    return df

def display_sorted_data(df, sort_by):
    if sort_by not in df.columns:
        raise ValueError(f"Invalid parameter '{sort_by}'. Choose from {df.columns.tolist()}.")

    sorted_df = df.sort_values(by=sort_by)
    print(f"Data sorted by {sort_by}:")
    print(sorted_df.to_string(index=False))

def main():
    num_records = 100
    df = generate_model_data(num_records)

    print("Generated Data:")
    print(df.head())

    sort_by = input("Enter the parameter to sort by (Fever, Cold, Shivering, Weight Loss): ")
    try:
        display_sorted_data(df, sort_by)
    except ValueError as e:
        print(e)

if __name__ == "__main__":
    main()
```

```python
#Q6
import re
import math

def parse_equation(equation):
    equation = equation.replace(" ", "")
    equation = equation.lower().replace("x^2", "x2").replace("^2", "")

    pattern = r"([+-]?\d*\.?\d*)x2([+-]\d*\.?\d*)x([+-]\d*\.?\d*)=0"
    match = re.match(pattern, equation)

    if match:
        a_str, b_str, c_str = match.groups()

        a = float(a_str) if a_str not in ["", "+", "-"] else (1.0 if a_str in ["", "+"] else -1.0)
        b = float(b_str) if b_str not in ["", "+", "-"] else (1.0 if b_str in ["", "+"] else -1.0)
        c = float(c_str) if c_str not in ["", "+", "-"] else (0.0 if c_str == "" else float(c_str))
        return a, b, c
    else:
        raise ValueError("The equation format is incorrect. Use the form 'ax^2 + bx + c = 0'.")

def solve_quadratic(a, b, c):
    discriminant = b**2 - 4*a*c

    if discriminant < 0:
        return "No real roots"

    root1 = (-b + math.sqrt(discriminant)) / (2 * a)
    root2 = (-b - math.sqrt(discriminant)) / (2 * a)

    return root1, root2

def main():
    print("Enter the quadratic equation in the form 'ax^2 + bx + c = 0'")
    equation = input("Equation: ")

    try:
        a, b, c = parse_equation(equation)
    except ValueError as e:
        print(e)
        return

    if a == 0:
        print("Coefficient a cannot be zero for a quadratic equation.")
        return

    roots = solve_quadratic(a, b, c)

    if roots == "No real roots":
        print(roots)
    else:
        root1, root2 = roots
        print(f"The roots of the equation are: {root1} and {root2}")

if __name__ == "__main__":
    main()
```
